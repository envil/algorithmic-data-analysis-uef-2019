```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

S = [2, 6, 6, 3, 3, 3, 5, 5, 3, 4, 6, 3, 2, 3, 6, 4]
binned_S = []

for i in range(0, len(S), 4):
    binned_S.append(np.average(S[i:i + 4]))
    
print('Binned S of size = 4')
print(binned_S)

plt.plot(range(len(S)), S)
plt.plot(range(3, len(S), 4), binned_S, c='r')
plt.title('Binned S of size 4')
plt.show()
```

```python
moving_average_S = []

for i in range(len(S) - 4):
    moving_average_S.append(np.average(S[i:i + 4]))

print('Moving average S of size = 4')
print(moving_average_S)

plt.plot(range(len(S)), S)
plt.plot(range(4, len(S)), moving_average_S, c='r')
plt.title('Moving average S of size 4')
plt.show()
```

```python
smoothed_S = []
ALPHA = 0.6
smoothed_S.append(S[0])
for i in range(1, len(S)):
    smoothed_S.append(ALPHA * S[i] + (1-ALPHA) * smoothed_S[i-1])

print('Smoothed S of size = 4')
print(smoothed_S)

plt.plot(range(len(S)), S)
plt.plot(range(len(S)), smoothed_S, c='r')
plt.title('Smoothed S of size 4')
plt.show()
```
